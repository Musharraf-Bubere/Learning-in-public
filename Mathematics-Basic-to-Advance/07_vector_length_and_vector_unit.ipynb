## üìò Introduction

A **vector** is a mathematical object that has both **magnitude (length)** and **direction**.  
Vectors are fundamental in **linear algebra**, **machine learning**, **physics**, and **data science**.

A vector can be represented in **2D, 3D, or n-dimensional space**.

---

## üìê Vector Representation

A vector in n-dimensional space is written as:

\[
\vec{x} =
\begin{bmatrix}
x_1 \\
x_2 \\
\vdots \\
x_n
\end{bmatrix}
\]

Examples:
- 2D vector ‚Üí \((x, y)\)
- 3D vector ‚Üí \((x_1, x_2, x_3)\)

---

## üìè Vector Length (Magnitude)

The **length (magnitude)** of a vector tells us how far the vector is from the origin.

### üîπ Formula

For vector:

\[
\vec{x} = (x_1, x_2, \dots, x_n)
\]

Magnitude is:

\[
\|\vec{x}\| = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
\]

---

## üßÆ Example

Given vector:

\[
\vec{A} = (2, 3)
\]

Magnitude:

\[
\|\vec{A}\| = \sqrt{2^2 + 3^2}
= \sqrt{13}
\]

This represents the **distance from the origin to the point (2,3)**.

## üìå Properties of Vector Length

- Always **non-negative**
- Represents **distance**
- Larger magnitude ‚Üí farther from origin
- Used in similarity, distance, and optimization problems

## üéØ Unit Vector

A **unit vector** is a vector whose magnitude is **exactly 1**.

\[
\|\hat{u}\| = 1
\]

Unit vectors represent **direction only**.

---

## üîÑ Converting a Vector into a Unit Vector

To convert any vector \(\vec{v}\) into a unit vector:

\[
\hat{u} = \frac{\vec{v}}{\|\vec{v}\|}
\]

This process is called **normalization**.

---

## üßÆ Example: Unit Vector

Given:

\[
\vec{v} = (1, 2, 0)
\]

Magnitude:

\[
\|\vec{v}\| = \sqrt{1^2 + 2^2 + 0^2} = \sqrt{5}
\]

Unit vector:

\[
\hat{u} =
\left(
\frac{1}{\sqrt{5}},
\frac{2}{\sqrt{5}},
0
\right)
\]

---

## ‚úÖ Verification

\[
\sqrt{
\left(\frac{1}{\sqrt{5}}\right)^2 +
\left(\frac{2}{\sqrt{5}}\right)^2
}
= 1
\]

So this is a **valid unit vector**.

---

## üìä Why Unit Vectors Are Important

- Used in **feature scaling**
- Used in **cosine similarity**
- Improves **gradient descent**
- Prevents dominance of large-scale features

---

## üß† Key Takeaways

- Vector length = magnitude of vector
- Unit vector has magnitude = 1
- Unit vectors preserve direction
- Normalization is critical in ML & optimization
