## ğŸ”¹ Linear Equations

A **linear equation** looks like:

\[
ax + by = c
\]

Multiple linear equations can be written in **matrix form**:

\[
AX = B
\]

Where:
- \(A\) â†’ coefficient matrix
- \(X\) â†’ unknown variables
- \(B\) â†’ output vector

This representation makes computation efficient.
