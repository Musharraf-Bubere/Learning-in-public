## Page 1: What is a Vector?

A **vector** is a mathematical quantity that has **both magnitude and direction**.

Unlike a scalar (which has only magnitude), a vector tells us:
- **How much**
- **In which direction**

This idea is commonly used in:
- Physics (force, velocity)
- Mathematics
- Machine Learning (data representation)

In machine learning, **each data record is treated as a vector**.

## Page 2: Scalar vs Vector

### Scalar
A **scalar** is a single numerical value.

Examples:
- Temperature = 30
- Speed = 60
- Learning rate = 0.01

A scalar has **only magnitude**, no direction.

---

### Vector
A **vector** has:
- Magnitude
- Direction

Example:
\[
(3,4)
\]

This means movement in a particular direction in space.

## Page 3: Vector Representation

Vectors are represented in a **coordinate system**.

A vector:
- Starts from the **origin**
- Ends at a point `(x, y)`

This representation helps us:
- Visualize vectors
- Perform mathematical operations
- Understand direction and distance

In data science, vectors represent **points in feature space**.

## Page 4: Magnitude of a Vector

The **magnitude** of a vector represents its **length**.

Formula:
\[
||v|| = \sqrt{x^2 + y^2}
\]

Example:
\[
(3,4) \Rightarrow \sqrt{9+16} = 5
\]

Magnitude is important for:
- Measuring distance
- Normalizing data
- Scaling features

## Page 5: Direction of a Vector

The **direction** of a vector tells us where it points.

Two vectors may:
- Have the same magnitude but different directions
- Have different magnitudes but the same direction

In machine learning:
- **Direction captures meaning**
- Used heavily in similarity measures

## Page 6: Vector Addition

Vectors are added **component-wise**.

\[
(x_1, y_1) + (x_2, y_2) = (x_1+x_2, y_1+y_2)
\]

Vector addition is used in:
- Physics (combined force)
- Machine learning (weight updates)

Example:
\[
w_{new} = w_{old} + \Delta w
\]

## Page 7: Scalar Multiplication

Scalar multiplication changes the **magnitude** of a vector.

\[
c(x, y) = (cx, cy)
\]

- Direction remains the same
- Length increases or decreases

Used in:
- Feature scaling
- Learning rate control
- Normalization

## Page 8: Dot Product

The **dot product** converts two vectors into a **scalar**.

\[
A \cdot B = \sum a_i b_i
\]

The dot product measures:
- Alignment
- Similarity

This is widely used in:
- Linear Regression
- Classification
- Recommendation systems
- NLP embeddings

## Final Summary

- Scalars represent single values
- Vectors represent multi-dimensional data
- Magnitude = size
- Direction = meaning
- Vector operations power machine learning
