## Overview

In this course, we will focus on **three core mathematical pillars** that form the foundation of **Data Science, Machine Learning, and Deep Learning**.

These concepts are essential not just for theory, but for **real-world applications**.

---

## The 3 Main Pillars

### 1. Linear Algebra

Linear Algebra is the backbone of **Machine Learning and Neural Networks**.

Topics covered include:
- Scalars and Vectors
- Vector Operations
- Matrices and Matrix Operations
- Linear Transformations
- Inverse Functions
- Eigenvalues and Eigenvectors

**Why it matters:**
- Neural Networks use **matrix multiplication**
- Forward & Backward propagation are matrix operations
- Dimensionality reduction (PCA) depends on eigenvalues & eigenvectors

---

### 2. Statistics (Basics to Advanced)

Statistics helps us **understand data and build models**.

Used extensively in:
- Machine Learning
- Deep Learning
- Model evaluation
- Large datasets

Statistics is divided into **two main types**:

#### a) Descriptive Statistics
Used to **summarize and understand data**
- Measure of Central Tendency (Mean, Median, Mode)
- Measure of Dispersion (Variance, Standard Deviation)
- Histograms and Box Plots
- Types of Data Distribution
- PMF, PDF, Normal Distribution, Skewness

#### b) Inferential Statistics
Used to **draw conclusions from data**
- Hypothesis Testing
- P-value
- Z-test and T-test
- Chi-square Test
- ANOVA Test (used for multiple comparisons)

### 3. Differential Calculus

Calculus plays a key role in **optimization problems**.

Topics include:
- Derivatives and Slopes
- Tangent Lines
- Polynomial Expressions
- Trigonometric Functions
- Chain Rule
- Composite Functions

**Why it matters:**
- Optimization uses gradients
- Gradient Descent is based on derivatives
- Backpropagation uses the chain rule

---

## Applications in Data Science

The combination of **Linear Algebra + Statistics + Calculus** enables:

1. **Linear Regression**
   - Simple Linear Regression
   - Multiple Linear Regression

2. **Dimensionality Reduction**
   - Principal Component Analysis (PCA)
   - Uses eigenvalues and eigenvectors

3. **Neural Networks**
   - ANN (Artificial Neural Networks)
   - Multi-layered architectures
   - Optimization using calculus

---

## Key Takeaways

- Linear Algebra enables efficient computation
- Statistics helps in understanding and modeling data
- Calculus enables optimization and learning
- Together, they form the foundation of **Data Science & AI**
