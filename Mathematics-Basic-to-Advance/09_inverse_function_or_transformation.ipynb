## ðŸ“Œ Page 1: What is an Inverse Function?

An **inverse of a function** is a function that **reverses the effect** of the original function.

If a function `f` maps an element `x` from set `X` to an element `y` in set `Y`, then the inverse function `fâ»Â¹` maps `y` **back to `x`**.

### Definition
If  
\[
f: X \rightarrow Y
\]
then the inverse function is:
\[
f^{-1}: Y \rightarrow X
\]

Such that:
\[
f(x) = y \Rightarrow f^{-1}(y) = x
\]

---

## ðŸ” Properties of Inverse Function

The inverse function satisfies:

1. **Forward then backward returns original value**
\[
f(f^{-1}(y)) = y
\]

2. **Backward then forward returns original value**
\[
f^{-1}(f(x)) = x
\]

These conditions guarantee **perfect reversibility**.

### Identity Function Definition

The **identity function** maps every element to itself:

\[
I_X: X \rightarrow X
\]

\[
I_X(a) = a \quad \forall a \in X
\]

### Matrix Representation
For vectors:
\[
I =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
\]

Multiplying any vector by `I` returns the **same vector**.

---

3. **Identity matrix** has 1s on diagonal, 0 elsewhere  
4. **Inverse of identity = identity**
