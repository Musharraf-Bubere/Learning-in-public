## Introduction to Bernoulli Distribution

The **Bernoulli distribution** is the **simplest discrete probability distribution**.  
It models a **random experiment with exactly two possible outcomes**:

- **Success** (with probability `p`)
- **Failure** (with probability `1 − p`)

This distribution is widely used to represent **binary outcomes**, such as:
- Coin toss (Head / Tail)
- Pass / Fail
- Yes / No
- Click / No Click

---

## Key Characteristics

- **Type**: Discrete Random Variable
- **Outcomes**: Binary → `{0, 1}`
- **Probability Mass Function (PMF)**

Let random variable **X** be:

- `X = 1` → Success
- `X = 0` → Failure

Then:

\[
P(X = 1) = p
\]

\[
P(X = 0) = 1 - p = q
\]

Where:

- \( 0 \le p \le 1 \)
- \( q = 1 - p \)

---

## Example 1: Tossing a Coin

Let:

- Success = Head
- Failure = Tail

\[
P(X = H) = 0.5
\]

\[
P(X = T) = 0.5
\]

This is a **Bernoulli trial** because there are only **two outcomes**.

---

## Example 2: Pass / Fail Scenario

Suppose:

- Probability of passing an exam = `0.4`
- Probability of failing = `0.6`

\[
P(X = \text{Pass}) = 0.4
\]

\[
P(X = \text{Fail}) = 0.6
\]

---

## Bernoulli PMF Formula

\[
P(X = k) = p^k (1 - p)^{1 - k}
\]

Where:

- \( k \in \{0, 1\} \)
- `k = 1` → Success
- `k = 0` → Failure

---

## Mean (Expected Value)

\[
E(X) = \sum k \cdot P(k)
\]

For Bernoulli distribution:

\[
E(X) = p
\]

---

## Median of Bernoulli Distribution

\[
\text{Median} =
\begin{cases}
0 & \text{if } p < 0.5 \\
[0,1] & \text{if } p = 0.5 \\
1 & \text{if } p > 0.5
\end{cases}
\]

---

## Mode of Bernoulli Distribution

- If \( p > q \) → **Mode = 1**
- If \( q > p \) → **Mode = 0**
- If \( p = q \) → Both are modes

---

## Variance of Bernoulli Distribution

\[
\text{Var}(X) = p(1 - p)
\]

\[
\sigma = \sqrt{p(1 - p)}
\]

## Summary

- Bernoulli distribution models **binary outcomes**
- Only **one parameter**: probability `p`
- Used in:
  - Probability theory
  - Machine Learning
  - Logistic Regression
  - Naive Bayes

---

## Key Takeaways

- Bernoulli is the **foundation of all discrete distributions**
- Outcomes are only `{0, 1}`
- Mean = `p`
- Variance = `p(1 − p)`
- Extremely important in **data science & ML**
