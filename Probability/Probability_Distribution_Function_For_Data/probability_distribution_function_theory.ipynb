## Introduction

**Probability Distribution Functions** describe how probabilities are distributed over the values of a **random variable**.

A random variable can be:
- **Discrete** → countable values  
- **Continuous** → values over an interval  

These distributions form the **foundation of probability, statistics, and machine learning**.

---

## Types of Probability Distribution Functions

There are **three main types** of probability distribution functions:

1. **Probability Mass Function (PMF)**  
2. **Probability Density Function (PDF)**  
3. **Cumulative Distribution Function (CDF)**  

---

## 1. Probability Mass Function (PMF)

> Used for **Discrete Random Variables**

A **PMF** gives the probability that a discrete random variable takes a **specific value**.

### Example: Rolling a Fair Dice

Let  
\[
X = \{1,2,3,4,5,6\}
\]

Since the dice is fair:

\[
P(X = x) = \frac{1}{6} \quad \text{for each } x
\]

So,

\[
P(1)=P(2)=P(3)=P(4)=P(5)=P(6)=\frac{1}{6}
\]

---

### PMF Visualization

![PMF Plot](image/01_PDF_PMF_CDF_Page_1.png)

Each bar represents the probability of a **single outcome**.

---

### Cumulative Probability from PMF

\[
P(X \le 2) = P(X=1) + P(X=2)
\]

\[
= \frac{1}{6} + \frac{1}{6} = \frac{1}{3}
\]

---

## 2. Cumulative Distribution Function (CDF)

> Used for **both discrete and continuous variables**

A **CDF** gives the probability that a random variable is **less than or equal to a value**.

\[
F(x) = P(X \le x)
\]

---

### CDF from Dice Example

\[
P(X \le 6) = 1
\]

Because all outcomes are ≤ 6.

---

### CDF Visualization

![CDF Plot](image/01_PDF_PMF_CDF_Page_1.png)

- CDF is **non-decreasing**
- Final value is always **1**

---

## 3. Probability Density Function (PDF)

> Used for **Continuous Random Variables**

Examples:
- Age
- Height
- Weight
- Salary
- Time

---

### Key Concept

- **Probability is area under the curve**
- Probability at a **single point is zero**

\[
P(X = x) = 0
\]

---

### PDF Example

![PDF Plot](image/01_PDF_PMF_CDF_Page_2.png)

\[
P(a \le X \le b) = \int_a^b f(x)\,dx
\]

---

## Relationship Between PDF and CDF

![PDF and CDF Relationship](image/01_PDF_PMF_CDF_Page_2.png)

\[
\text{PDF} = \frac{d}{dx}(\text{CDF})
\]

\[
f(x) = \frac{dF(x)}{dx}
\]

---

## Properties of Probability Density Function (PDF)

1. **Non-negativity**
\[
f(x) \ge 0
\]

2. **Total area under curve = 1**
\[
\int_{-\infty}^{\infty} f(x)\,dx = 1
\]

## Different Types of Probability Distributions

![Distribution Types](image/01_PDF_PMF_CDF_Page_3.png)

### Common Distributions

1. **Bernoulli Distribution** → PMF  
2. **Binomial Distribution** → PMF  
3. **Poisson Distribution** → PMF  
4. **Normal (Gaussian) Distribution** → PDF  
5. **Log-Normal Distribution** → PDF  
6. **Uniform Distribution** → PMF / PDF

## Summary

- **PMF** → discrete random variables  
- **PDF** → continuous random variables  
- **CDF** → cumulative probability  
- PDF is the **derivative of CDF**
- Area under PDF curve equals **1**

---

## Key Takeaways

- Probability distributions explain how data behaves
- Discrete → PMF
- Continuous → PDF
- CDF accumulates probability
- Core foundation for **statistics, ML, and data science**
