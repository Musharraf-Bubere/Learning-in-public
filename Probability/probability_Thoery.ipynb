## Introduction to Probability

**Probability** is a measure of how likely an event is to occur.

It plays a crucial role in:
- Statistics
- Machine Learning
- Classification algorithms (threshold-based decisions)

---

## What is Probability?

Probability tells us the likelihood of an experiment or event happening.

### Example 1: Tossing a Coin

Possible outcomes:
\[
\{H, T\}
\]

\[
P(H) = \frac{1}{2} = 50\%
\]
\[
P(T) = \frac{1}{2} = 50\%
\]

---

### Example 2: Rolling a Dice

Possible outcomes:
\[
\{1, 2, 3, 4, 5, 6\}
\]

Probability of getting any number:
\[
P(X = 1) = \frac{1}{6}
\]

---

## Mutually Exclusive Events

Two events are **mutually exclusive** if they **cannot occur at the same time**.

### Examples
- Head and Tail in a coin toss
- Rolling 1 and 5 on a dice (cannot happen together)

### Visual Idea
- Venn diagram **does not overlap**

---

### Addition Rule (Mutually Exclusive)

\[
P(A \text{ or } B) = P(A) + P(B)
\]

---

### Example 1: Coin Toss

\[
P(H \text{ or } T) = \frac{1}{2} + \frac{1}{2} = 1
\]

---

### Example 2: Dice Roll

Probability of getting **1 or 5**:

\[
P(1 \text{ or } 5) = \frac{1}{6} + \frac{1}{6} = \frac{2}{6} = \frac{1}{3}
\]

---

## Non-Mutually Exclusive Events

Events are **non-mutually exclusive** if they **can occur together**.

### Example: Drawing a Card from a Deck

- Event A: Drawing a **King**
- Event B: Drawing a **Heart**

ðŸ‘‰ **King of Hearts** satisfies both events.

---

### Addition Rule (Non-Mutually Exclusive)

\[
P(A \text{ or } B) = P(A) + P(B) - P(A \text{ and } B)
\]

---

### Example: Deck of Cards (52 cards)

| Event | Probability |
|------|------------|
| King | \( \frac{4}{52} \) |
| Heart | \( \frac{13}{52} \) |
| King of Hearts | \( \frac{1}{52} \) |

\[
P(K \text{ or } H) = \frac{4}{52} + \frac{13}{52} - \frac{1}{52}
\]

\[
= \frac{16}{52}
\]

---

## Summary

- Probability measures likelihood of events
- Addition rule depends on whether events overlap
- Overlapping probability must be **subtracted once**

---

## Key Takeaways

- **Mutually exclusive** â†’ no overlap  
  \[
  P(A \text{ or } B) = P(A) + P(B)
  \]

- **Non-mutually exclusive** â†’ overlap exists  
  \[
  P(A \text{ or } B) = P(A) + P(B) - P(A \cap B)
  \]

- Addition rules are fundamental in probability theory and ML

## Introduction

In this session, we continue our discussion on **probability**.  
Previously, we studied the **addition rule** for mutually exclusive and non-mutually exclusive events.  
Now, we focus on the **multiplication rule**, which is used for:

- **Independent events**
- **Dependent events**

These concepts are fundamental in **statistics, machine learning, and Naive Bayes algorithms**.

---

## Independent Events

Two events are **independent** if the occurrence of one **does not affect** the probability of the other.

### Example: Tossing a Coin

- Probability of getting **Head**:
\[
P(H) = \frac{1}{2}
\]

- Probability of getting **Tail**:
\[
P(T) = \frac{1}{2}
\]

Even if the first toss is Head, the probability of Tail in the second toss remains the same.  
Hence, coin tosses are **independent events**.

---

### Example: Rolling a Dice

- Probability of getting any specific number:
\[
P(X = 1) = \frac{1}{6}
\]

Each roll is independent of the previous roll.

---

## Multiplication Rule for Independent Events

For independent events:

\[
P(A \text{ and } B) = P(A) \times P(B)
\]

### Example

Probability of getting **Head and then Tail**:

\[
P(H \text{ and } T) = \frac{1}{2} \times \frac{1}{2} = \frac{1}{4}
\]

---

![Independent Events](images/02_Probability_Multiplication_Rule_Page_2.png)

---

## Dependent Events

Two events are **dependent** if the occurrence of one event **affects** the probability of the other.

---

### Example: Drawing Cards from a Deck

- Total cards = 52
- Number of Kings = 4

#### First Event:
\[
P(K) = \frac{4}{52}
\]

After drawing a King, the deck now has **51 cards**.

#### Second Event (Queen given King is drawn):
\[
P(Q \mid K) = \frac{4}{51}
\]

---

## Multiplication Rule for Dependent Events

For dependent events:

\[
P(A \text{ and } B) = P(A) \times P(B \mid A)
\]

### Example

\[
P(K \text{ and } Q) = \frac{4}{52} \times \frac{4}{51}
\]

This concept is called **conditional probability** and is the foundation of **Bayesâ€™ Theorem**.

---

## Summary

- **Independent events** do not influence each other  
- **Dependent events** influence subsequent probabilities  
- Multiplication rule differs based on dependency  
- Conditional probability is crucial in **ML algorithms**

---

## Key Takeaways

- Independent Events:
\[
P(A \text{ and } B) = P(A) \times P(B)
\]

- Dependent Events:
\[
P(A \text{ and } B) = P(A) \times P(B \mid A)
\]

- Used extensively in:
  - Probability theory
  - Bayesâ€™ theorem
  - Naive Bayes classifier
