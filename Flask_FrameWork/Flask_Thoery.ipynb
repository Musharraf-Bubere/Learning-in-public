#### Flask :- Web framework which is created with the help of Python Programming Language

### 01. WSGI --> Web Server Gateway Interface
![WSGI Architecture](images/01_WSGI.jpeg)

***Explanation:***
- Client sends multiple requests (Req1, Req2, Req3)
- Requests first reach the *Web Server* (Apache, IIS, AWS EC2, Azure)
- *WSGI* acts as a bridge between the Web Server and Flask Web App
- Flask handles routes like /home, /api, /functionality
- Response is sent back to the client via the Web Server

### 02. Jinja 2 Template Engine
![Jinja2 Template Engine Flow](images/02_Jinja_2_Template_Engine.jpeg)

**Explanation:**
- Jinja2 is a web template engine used by Flask
- It connects web templates with dynamic data sources
- Data can come from SQL DB, CSV files, ML models, or MongoDB
- The rendered output is sent as an HTML web page
