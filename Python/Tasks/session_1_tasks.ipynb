### Q2:- Write a program that will convert celsius value to fahrenheit.

````python
# Write your code here
c = float(input("Enter Celsius Number: "))
fahrenheit = (c*1.8) + 32
print(fahrenheit)
````

````python
# Cleaner Way using f-string
# After Evaluate

c = float(input("Enter Celsius value: "))
print(f"Fahrenheit: {(c * 1.8) + 32}")
````

### Q3:- Take 2 numbers as input from the user.Write a program to swap the numbers without using any special python syntax.

````python
# Write your code here
num1 = int(input("Enter first number:"))
num2 = int(input("Enter second number:"))

print(num1)
print(num2)
print("Swap the number:", num2, num1)
````

````python
# Correct Approach (Without Special Python Syntax)
# After Evaluate

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

temp = num1
num1 = num2
num2 = temp

print("After swapping:")
print("num1 =", num1)
print("num2 =", num2)
````

### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
import math
# Write your code here
x1 = float(input("Enter coordinate x1:"))
x2 = float(input("Enter coordinate x2:"))
y1 = float(input("Enter coordinate y1:"))
y2 = float(input("Enter coordinate y2:"))

d = math.sqrt((x2-x1)**2+(y2-y1)**2)
print(d)
````

### Q5:- Write a program to find the simple interest when the value of principle,rate of interest and time period is provided by the user.
