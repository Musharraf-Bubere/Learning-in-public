# Tuples

A tuple in python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned wheereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterestics

- Ordered
- Unchangeble
- Allows duplicate

### Plan of attack

- Creating a Tuple
- Accessing items
- Editing items
- Adding items
- Deleting items
- Operations on Tuples
- Tuple Functions

```python
# Creating Tuples

# empty
t1 = ()
print(t1)

# create a tuple with a single item
t2 = (2,)
print(t2)
print(type(t2))

# Homogeneous
t3 = (1,2,3,4)
print(t3)

# Hetro
t4 = (1,2,2.5,True,[6,7])
print(t4)

# tuple
t5 = (1,2,3,(4,5))
print(t5)

# using type conversion
t6 = tuple('hello')
print(t6)
```

### Accssing Items

- Indexing
- Slicing

```python
# Indexing

# positive indexing
print(t3)
print(t3[0])

# Negative indexing
print(t3[-1])

# Slicing
print(t3[0:4:2])

print(t5)
print(t5[-1][0])
```

```python
# Editing items

print(t3)
t3[0] = 100
# immutable just like strings
```

```python
# Adding items
print(t3)
# not possible
```

```python
# Deleting items

# print(t3)
# del t3
# # print(t3)

print(t5)
del t5[-1]
print(t5)
```

```python
# Operations on Tuple

# + and *
t1 = (1,2,3,4)
t2 = (5,6,7,8)
print(t1 + t2)
print(t1 *3)

# Membership
1 in t1

# Iteration
for i in t1:
    print(i)
```

### Difference between Lists and Tuples

- Syntax
- Mutability
- Speed 
- Memory
- Built in Functionality
- Error prone
- Usability

```python
import time

l = list(range(100000000))
t = tuple(range(100000000))

start = time.time()
for i in l:
    i*5
print("List time",time.time()-start)

start = time.time()
for i in t:
    i*5
print("Tuple time",time.time()-start)
```

```python
import sys

l = list(range(1000))
t = tuple(range(1000))

print("List size",sys.getsizeof(l))
print("Tuple size",sys.getsizeof(t))
```

# Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed) However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characteristics:

- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

### Creating Sets

```python
# empty
s = set()
print(s)
print(type(s))

# 1D and 2D
s1 = {1,2,3}
print(s1)

# s2 = {1,2,3,{4,5}}
# print(s2)

# Homo and Hetro
s3 = {1,"Hello",2.5,True,(1,2,3)}
print(s3)

# Using type conversion
s4 = set([1,2,3,4])
print(s4)

# duplicates not allowed
s5 = {1,2,3,5,1,1}
print(s5)

# Set can't have mutable items
s6 = {1,2,[3,4]}
print(s6)
```

```python
s1 = {1,2,3}
s2 = {3,2,1}

print(s1 == s2)
```

### Accessing Items

```python
s1 = {1,2,3,4}
s1[0]
```

### Editing Items

```python
s1 = {1,2,3,4}
s1[0] = 100
```

### Adding Items

```python
s = {1,2,3,4}
# add
s.add(5)
print(s)

# update
s.update([6,7,8])
print(s)
```

### Deleting Items

```python
# del
s = {1,2,3,4,5}
# print(s)
# del s
# print(s)

# discard
s.discard(5)
print(s)

# remove
s.remove(3)
print(s)

# pop
s.pop()
print(s)

# clear
s.clear()
print(s)
```

### Set Operations

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# union(|)
print(s1 | s2)

# intersection(&)
print(s1 & s2)

# Difference(-)
print(s1 - s2)
print(s2 - s1)

# Symmetric Difference(^)
print(s1 ^ s2)

# Membership Test
print(1 in s1)
print(1 not in s1)

# Iteration
for i in s1:
    print(i)
```

# Set Functions

```python
# len/sum/min/max/sorted
s = {3,1,4,3,5,}
len(s)
sum(s)
min(s)
max(s)
sorted(s,reverse=True)
```

```python
# union/update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 | s2
s1.union(s2)

s1.update(s2)
print(s1)
print(s2)
```

```python
# intersection/intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 & s2
print(s1.intersection(s2))

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 & s2
print(s1.difference(s2))

s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 & s2
print(s1.symmetric_difference(s2))

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/ussuperset
s1 = {1,2,3,4,5}
s2 = {4,5}

print(s1.isdisjoint(s2))
print(s1.issubset(s2))
print(s1.issuperset(s2))
```

```python
# copy
s1 = {1,2,3,4}
s2 = s1.copy()
print(s1)
print(s2)
```

### Dictionary

Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is know as map or assosiative arrays.

dict = {'name': 'nitish', 'age':33, 'gender': 'male'}

Characteristics

- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items

### Create Dictionary

```python
# Empty dictionary
d = {}
print(d)

# 1D dictionary
d1 = {'name': 'nitish','gender': 'male'}
print(d1)

# with mixed keys
d2 = {(1,2,3):1, "hello":"world"}
print(d2)

# 2D dictionary -> JSON
s = {
    "name": "musharraf",
    "college": "bit",
    "sem": 4,
    "subjects":{
        "dsa":50,
        "maths":67,
        "english":34
    }
}
print(s)

# using squence and dict function
d4 = dict([(1,1),(2,2),(3,3)])
print(d4)

# duplicate keys
d5 = {'name':'nitish','name':'rahul'}
print(d5)

# mutable items as keys
# d6 = {'name':'nitish',[1,2,3]:2}
# print(d6)
```

### Accessing Items

```python
my_dict = {'name': 'Jack', 'age': 26}

# []
my_dict['age']

# get
my_dict.get('name')

print(s)
s['subjects']['maths']
```

### Adding key-value pair

```python
print(d1)

d1['city'] = 'Mumbai'
print(d1)

d1['weight'] = 72
print(d1)

print(s)
s['subjects']['ds'] = 75
print(s)
```

### Remove key-value pair

### Editing key-value pair

```python
s['sem'] = 5
print(s)

s['subjects']['dsa'] = 80
print(s)
```

### Dictionary Comprehension

```python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
```

```python
# using existing dict
distances = {'delhi':1000,'mumbai':2000,'bangalore':3000}

{key:value*0.62 for (key,value) in distances.items()}
```

```python
# using zip
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
temp_c = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{i:j for (i,j) in zip(days,temp_c)}
```

```python
# using if condition
products = {'phone':10, 'laptop':0, 'charger':32, 'tablet':0}

{key:value for (key,value) in products.items() if value > 0}
```

```python
# Nested Comprehension
# print tables of number from 2 to 4

{i:{j:i*j for j in range(1,11)}for i in range(2,5)}
```
