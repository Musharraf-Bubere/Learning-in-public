# Operators in Python
- Arithmatic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

```python
# Arithmatic Operators
print(5+6)

print(5-6)

print(5*6)

print(5/2)

print(5//2)

print(5%2)

print(5**2)
```

```python
# Relational Operators
print(4>5)

print(4<5)

print(4>=5)

print(4<=5)

print(4==5)

print(4!=5)
```

```python
# Logical Operators
print(1 and 0)

print(1 or 0)

print(not 1)
```

```python
# Bitwise Operators

# bitwise and
print(2 & 3) 

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# bitwise not
print(~ 3)

# bitwise leftshift
print(4 >> 2)

# bitwise rightshift
print(5 << 2)
```

```python
# Assignment Operators

# =
a = 2

# +=
a += 2

# -=
a -= 2

# *=
a *= 2

# /=
a /= 2
```

```python
# Membership Operators

# in/not in
print('D' in 'Delhi')
print('D' not in 'Delhi')

print(1 in [2,3,5,6])
```

# if-else in Python

```python
# login program and indentation
# email -> nitish.campusx@gmail.com
# password -> 1234

email = input("Enter email:")
password = input("Enter password:") 

if email == "nitish.campusx@gmail.com" and password == '1234':
    print("Welcome")
elif email == "nitish.campusx@gmail.com" and password != '1234':
    # tell the user
    print("Incorrect Password")
    password = input("Enter password again")
    if password == "1234":
        print("Welcome, finally")
    else:
        print("Beta tumse na ho payenga!")
else:
    print("Not Correct")
```

```python
# if-else examples
# 1. Find the min of 3 given numbers
# 2. Menu Driven Program

# min of 3 numbers
a = int(input("First num:"))
b = int(input("Second num:"))
c = int(input("Third num:"))

if a < b and a < c:
    print("Smallest is:", a)
elif b < c:
    print("Smallest is:", b)
else:
    print("Smallest is", c)
```

```python
# Menu driven calculator
fnum = int(input("Enter the first number:"))
snum = int(input("Enter the second number:"))

op = input("Enter the operation(+,-,*,/):")

if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum * snum)
else:
    print(fnum / snum)
```

```python
# While loop example -> program to print the table
# program -> Sum of all digits of a given number
# program -> keep accepting numbers from users till he/she enters a 0 and then find the avg
```

```python
# Program to print the table
number = int(input("Enter the number"))

i = 1

while i < 11:
    print(number, "*", i, "=", number * i)
    i += 1 
```

```python
# while loop with else

x = 1

while x < 3:
    print(x)
    x += 1

else:
    print("limit crossed")
```

```python
# Guessing game

# generate a random integer between 1 and 100
import random 
jackpot = random.randint(1,100)

guess = int(input("Guess number:"))

counter = 1

while guess != jackpot:
    if guess < jackpot:
        print("Wrong! guess higher")
    else:
        print("Wrong! guess lower")

    guess = int(input("Guess number:"))
    counter += 1
else:
    print("Correct guess")
    print("Attempts", counter)
    print("jackpot Number", jackpot)
```

### Squence Sum

```python
# Code here

n = int(input("Enter n:"))

result = 0
fact = 1

for i in range(1,n+1):
    fact = fact * i
    result = result + i/fact

print(result)
```

### Pattern 1

```python
# Code here

rows = int(input("Enter number of rows:"))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print("*",end="")
    print()
```

### Pattern 2

1

121

12321

1234321

```python
# Code here

rows = int(input("Enter number of rows:"))

for i in range(1,rows+1):
    for j in range(1,i+1):
        print(j,end="")
    for k in range(i-1,0,-1):
        print(k,end="")
    print()
```

# Loop control Statments
- Break
- Continue
- Pass

```python
# Break

for i in range(1,10):
    if i == 5:
        break
    print(i)
```

```python
# Prime number 

lower = int(input("Enter lower range:"))
upper = int(input("Enter upper range:"))

for i in range(lower,upper+1):
    for j in range(2,i):
        if i % j == 0:
            break
    else:
        print(i)
```

```python
# Continue
for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
# pass
for i in range(1,10):
    pass
```
