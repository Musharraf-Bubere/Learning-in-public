1. Question: Write a function that takes a list of numbers and returns the sum.

```python
def sum_nums(l1):
    return sum(l1)

l1 = [1,2,3,4]
print(sum_nums(l1))
```

2. Question: How do you reverse a string in Python?

```python
s1 = "hello world"
s1 = list(s1)
s1.reverse()
for i in s1:
    print(i,end="")
```

```python
def reverse_string(s1):
    return s1[::-1]

s1 = "hello world"
print(reverse_string(s1))
```
