# 📈 Equation of Line, Plane and Hyperplane

---

# 🧠 Why this topic is important in Machine Learning

Many ML algorithms such as:

• Linear Regression  
• Logistic Regression  
• Support Vector Machines (SVM)

use **lines or planes to separate data**.

Example:

```
Spam vs Not Spam
Pass vs Fail
Cat vs Dog
```

The model tries to find a **best decision boundary**.

---

# 📊 1️⃣ Equation of a Straight Line (2D)

In 2D space we represent a line using:

```
y = mx + c
```

Where:

| Symbol | Meaning |
|------|------|
| m | slope of the line |
| c | y-intercept |

📌 **Slope (m)**  
Shows how steep the line is.

📌 **Intercept (c)**  
Point where the line crosses the **y-axis**.

---

# 📐 Alternative Representation

A line can also be written as:

```
ax + by + c = 0
```

Rearranging:

```
by = -ax - c
y = (-a/b)x - c/b
```

Comparing with:

```
y = mx + c
```

We get:

```
m = -a/b
intercept = -c/b
```

---

# 📊 ML Representation of Line

In Machine Learning we write the equation as:

```
y = β₀ + β₁x
```

Where:

| Term | Meaning |
|------|------|
| β₀ | intercept |
| β₁ | weight / slope |

---

# 📈 2️⃣ Line in Feature Space

If we have **two features**

```
x₁ , x₂
```

The equation becomes:

```
w₁x₁ + w₂x₂ + b = 0
```

Where:

| Symbol | Meaning |
|------|------|
| w₁ , w₂ | weights |
| b | bias |

This equation represents a **decision boundary**.

---

# 📦 Vector Representation

We can write it compactly as:

```
wᵀx + b = 0
```

Where:

```
w = weight vector
x = feature vector
```

Example:

```
w = [w₁ , w₂]
x = [x₁ , x₂]
```

---

# 📊 3️⃣ Plane in 3D

If we have **three features**

```
x₁ , x₂ , x₃
```

The equation becomes:

```
w₁x₁ + w₂x₂ + w₃x₃ + b = 0
```

This represents a **plane in 3D space**.

---

# 📐 4️⃣ Hyperplane (Higher Dimensions)

In machine learning datasets we often have many features.

Example:

```
x₁ , x₂ , x₃ , ... , xₙ
```

The equation becomes:

```
w₁x₁ + w₂x₂ + ... + wₙxₙ + b = 0
```

This is called a **Hyperplane**.

📌 Definition:

```
Hyperplane = Decision boundary in n-dimensional space
```

---

# 🎯 Geometric Interpretation

Dot product representation:

```
wᵀx = ||w|| ||x|| cosθ
```

Where:

| Symbol | Meaning |
|------|------|
| θ | angle between vectors |
| ||w|| | magnitude of w |
| ||x|| | magnitude of x |

If:

```
wᵀx = 0
```

Then:

```
cosθ = 0
θ = 90°
```

Meaning:

```
w ⟂ plane
```

The **weight vector is perpendicular to the hyperplane**.

---

# 📊 Special Case

Line passing through origin:

```
wᵀx = 0
```

Here:

```
b = 0
```

---

# 🧠 Quick Summary

| Concept | Equation |
|------|------|
| Line (2D) | y = mx + c |
| ML Line | y = β₀ + β₁x |
| Decision Boundary | w₁x₁ + w₂x₂ + b = 0 |
| Plane (3D) | w₁x₁ + w₂x₂ + w₃x₃ + b = 0 |
| Hyperplane | wᵀx + b = 0 |

---

# 🚀 Why this matters

Algorithms that use this concept:

• Linear Regression  
• Logistic Regression  
• Support Vector Machines  

They try to find the **best hyperplane that separates the data**.
