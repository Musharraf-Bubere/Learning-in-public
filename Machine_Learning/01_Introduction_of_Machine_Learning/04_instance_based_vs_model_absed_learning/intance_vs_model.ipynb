# 🤖 Instance Based Learning vs Model Based Learning

---

# 📌 Overview

Machine Learning models learn patterns from data in **two main ways**:

1️⃣ Instance Based Learning  
2️⃣ Model Based Learning

```
Learning
   │
   ├── Memorizing → Instance Based Learning
   │
   └── Generalizing → Model Based Learning
```

---

# 🧠 1️⃣ Instance Based Learning

📌 **Definition**

Instance based learning stores the **training data** and makes predictions by comparing new data with stored examples.

Instead of building a mathematical model, it **memorizes the dataset**.

---

## ⚙️ How it works

Steps:

1️⃣ Store the training data  
2️⃣ When a new data point arrives  
3️⃣ Compare it with similar data points  
4️⃣ Predict based on nearest examples

```
Training Data → Stored
New Input → Compare with stored data → Prediction
```

---

## 📊 Example

Dataset:

| Study Hours | Play Hours | Result |
|-------------|-----------|--------|
| 7 | 3 | Pass |
| 2 | 6 | Fail |

If a new student comes:

```
Study = 6
Play = 4
```

The algorithm finds **nearest neighbors** and predicts result.

---

## 🔹 Example Algorithm

```
KNN (K-Nearest Neighbours)
```

KNN finds the **closest K data points** and predicts the majority class.

---

# 🧠 2️⃣ Model Based Learning

📌 **Definition**

Model based learning creates a **generalized model** by learning patterns from the training data.

Instead of memorizing the dataset, it learns a **mathematical function**.

---

## ⚙️ How it works

Steps:

1️⃣ Train model on dataset  
2️⃣ Learn patterns from data  
3️⃣ Create a generalized model  
4️⃣ Use the model to predict new data

```
Training Data → Learn Pattern → Build Model → Prediction
```

---

## 📊 Example

Predict house price using:

```
Size of house
Number of rooms
Location
```

The algorithm learns a function like:

```
Price = f(Size, Rooms, Location)
```

Then it predicts prices for **new houses**.

---

# 📊 Decision Boundary Concept

Model based learning creates a **decision boundary** separating classes.

Example:

```
Pass students
Fail students
```

The model finds a boundary separating the two groups.

---

# 🔑 Key Difference

| Feature | Model Based Learning | Instance Based Learning |
|-------|---------------------|------------------------|
| Training | Train a model | No training model |
| Data storage | Model stored | Entire dataset stored |
| Prediction | Uses learned model | Uses training data directly |
| Speed | Fast prediction | Slower prediction |
| Memory usage | Low | High |

---

# 📦 Storage

Model based learning stores:

```
Model parameters
```

Instance based learning stores:

```
Entire training dataset
```

---

# 📊 Examples of Algorithms

### Model Based Learning

- Linear Regression
- Logistic Regression
- Decision Trees
- Random Forest
- Neural Networks

### Instance Based Learning

- KNN (K-Nearest Neighbours)

---

# 🧠 Quick Summary

✔ Instance Based Learning → **memorize data**

✔ Model Based Learning → **learn patterns**

✔ KNN → Instance based algorithm

✔ Linear Regression / Logistic Regression → Model based algorithms
