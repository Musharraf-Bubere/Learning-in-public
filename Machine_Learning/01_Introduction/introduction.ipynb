# 🤖 Introduction to Machine Learning
## AI vs ML vs DL vs Data Science

---

# 🧠 1. Artificial Intelligence (AI)

📌 **Definition**

Artificial Intelligence refers to building systems or applications that can perform tasks **without human intervention**.

These systems can:
- Think
- Learn
- Make decisions
- Solve problems automatically

🎯 **Goal of AI**
Create intelligent machines that behave like humans.

📍 **Examples**

• 🎬 Netflix Recommendation System  
   - If you watch **Action Movies**
   - Netflix recommends **More Action Movies**

• 🚗 Self Driving Cars

• 🗣️ Voice Assistants  
   - Alexa  
   - Siri  
   - Google Assistant

---

# ⚙️ 2. Machine Learning (ML)

📌 **Definition**

Machine Learning is a **subset of Artificial Intelligence** where machines **learn patterns from data instead of being explicitly programmed**.

Instead of writing rules manually, the system learns automatically from data.

📊 **Example**

Spam Email Detection

Traditional Programming:
```
if email contains "free money"
    mark as spam
```

Machine Learning:
- Train model with thousands of emails
- Model learns patterns automatically

📍 **Applications**

• Email Spam Detection  
• House Price Prediction  
• Fraud Detection  
• Movie Recommendation

---

# 🧬 3. Deep Learning (DL)

📌 **Definition**

Deep Learning is a **subset of Machine Learning** that uses **Artificial Neural Networks (ANN)** with multiple layers.

It tries to **mimic the human brain**.

🧠 **Key Idea**

```
Deep Learning = Multi Layer Neural Networks
```

📍 **Applications**

• Face Recognition  
• Image Classification  
• Speech Recognition  
• ChatGPT / LLM models

---

# 📊 4. Data Science (DS)

📌 **Definition**

Data Science is a field that focuses on **analyzing, visualizing, predicting, and extracting insights from data**.

It combines:

• Statistics  
• Machine Learning  
• Data Analysis  
• Data Visualization

📍 **Common Tools**

• Python  
• SQL  
• Excel  
• Power BI  
• Tableau

---

# 📈 Relationship Between AI, ML, DL, DS

```
Artificial Intelligence
        │
        ▼
Machine Learning
        │
        ▼
Deep Learning
```

📌 **Important**

• Machine Learning is a part of **AI**  
• Deep Learning is a part of **ML**  
• Data Science overlaps with ML because it uses ML algorithms
