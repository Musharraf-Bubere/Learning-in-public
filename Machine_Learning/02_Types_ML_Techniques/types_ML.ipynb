# 🤖 Types of Machine Learning Techniques

---

# 🧠 Overview

Machine Learning techniques are mainly divided into **3 types**:

1️⃣ Supervised Learning  
2️⃣ Unsupervised Learning  
3️⃣ Reinforcement Learning

```
Machine Learning
      │
 ┌────┼─────────────┐
 │    │             │
 ▼    ▼             ▼
Supervised   Unsupervised   Reinforcement
Learning      Learning        Learning
```

---

# 📘 1️⃣ Supervised Learning

📌 **Definition**

Supervised learning is a type of machine learning where the model learns from **labeled data**.

This means the dataset already contains the **correct output (target)**.

```
Input (X)  →  Model  →  Output (Y)
```

---

# 📊 Example: House Price Prediction

| Size of House | No. of Rooms | Price |
|---------------|--------------|------|
| 5000 | 5 | 450K |
| 6000 | 6 | 500K |

📌 **Independent Features (Input)**  
- Size of house  
- Number of rooms  

📌 **Dependent Feature (Output / Target)**  
- House price

The model learns the relationship between inputs and output.

---

# 🔹 Types of Supervised Learning Problems

Supervised learning problems are mainly of **two types**:

### 1️⃣ Regression

Used when the output variable is **continuous numeric value**.

📊 Examples

- House price prediction  
- Temperature prediction  
- Stock price prediction

Example output:

```
450000
500000
620000
```

---

### 2️⃣ Classification

Used when the output variable is **categorical**.

Example:

| Study Hours | Play Hours | Result |
|-------------|------------|--------|
| 7 | 3 | Pass |
| 2 | 6 | Fail |

📌 Output is **categorical**

Possible classes:

- Pass
- Fail

---

### Types of Classification

🔹 **Binary Classification**

Two classes

```
Pass / Fail
Spam / Not Spam
Yes / No
```

🔹 **Multiclass Classification**

More than two classes

```
Cat / Dog / Horse
Low / Medium / High
```

---

# 📘 2️⃣ Unsupervised Learning

📌 **Definition**

Unsupervised learning works with **unlabeled data**.

There is **no target variable**.

The algorithm tries to **find patterns or groups in the data**.

---

# 📊 Example: Customer Segmentation

Dataset:

| Salary | Spending Score |
|------|------|
| 20000 | 9 |
| 45000 | 2 |

The algorithm groups customers into **clusters**.

Example clusters:

```
Cluster 1 → High spending customers
Cluster 2 → Medium spending customers
Cluster 3 → Low spending customers
```

📌 Companies use this for:

- Marketing
- Recommendation systems
- Customer targeting

Example:

```
Ecommerce company sends same discount to customers in same cluster
```

---

# 🔹 Algorithms used in Unsupervised Learning

1️⃣ K-Means Clustering  

2️⃣ Hierarchical Clustering  

3️⃣ DBSCAN Clustering  

---

# 📘 3️⃣ Reinforcement Learning

📌 **Definition**

Reinforcement learning is a learning method where an **agent learns by interacting with an environment**.

The agent receives:

- 🎯 Reward (good action)
- ❌ Penalty (bad action)

Goal → **maximize reward**

---

# 📊 Examples of Reinforcement Learning

🚗 Self Driving Cars  

🎮 Game Playing AI  
- AlphaGo
- Chess AI

🤖 Robotics

---

# 📊 Summary

| Type | Data | Goal |
|-----|-----|-----|
| Supervised Learning | Labeled Data | Predict output |
| Unsupervised Learning | Unlabeled Data | Find patterns |
| Reinforcement Learning | Environment Interaction | Maximize reward |

---

# 🧠 Quick Revision

✔ Supervised Learning → Regression + Classification  

✔ Regression → Continuous output  

✔ Classification → Categorical output  

✔ Unsupervised Learning → Clustering  

✔ Reinforcement Learning → Reward based learning
