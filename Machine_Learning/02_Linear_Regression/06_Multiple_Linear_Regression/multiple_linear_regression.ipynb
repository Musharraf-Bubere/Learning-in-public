# 📈 Multiple Linear Regression

---

# 🧠 What is Multiple Linear Regression?

Multiple Linear Regression is an extension of **Simple Linear Regression** where we use **more than one input feature** to predict the output.

Example:

Predict **house price** using:

- Number of rooms
- Size of house
- Location

---

# 📊 Example Dataset

| Rooms (x₁) | Size (x₂) | Location (x₃) | Price (y) |
|------------|-----------|--------------|-----------|
| 2 | 1200 | City | 200000 |
| 3 | 1500 | City | 250000 |
| 4 | 1800 | Suburb | 320000 |

Where:

- **x₁, x₂, x₃ → Independent features**
- **y → Dependent variable**

---

# 📈 Hypothesis Function

In Multiple Linear Regression the hypothesis becomes:

```
hθ(x) = θ₀ + θ₁x₁ + θ₂x₂ + θ₃x₃
```

General form:

```
hθ(x) = θ₀ + θ₁x₁ + θ₂x₂ + ... + θₙxₙ
```

Where:

| Parameter | Meaning |
|----------|---------|
| θ₀ | Intercept |
| θ₁, θ₂, θ₃ | Coefficients (weights) |
| x₁, x₂, x₃ | Input features |

---

# 🧠 Interpretation

Each coefficient represents **importance of that feature**.

Example:

```
Price = θ₀ + θ₁(Rooms) + θ₂(Size) + θ₃(Location)
```

Meaning:

- θ₁ → effect of number of rooms
- θ₂ → effect of house size
- θ₃ → effect of location

---

# 📉 Cost Function

Same cost function as simple regression:

```
J(θ) = 1/(2m) Σ (hθ(xᵢ) − yᵢ)²
```

Goal:

```
Minimize J(θ)
```

---

# ⚙️ Optimization

We use **Gradient Descent** to update parameters.

Update rule:

```
θⱼ = θⱼ − α (∂J / ∂θⱼ)
```

Where:

- **α → learning rate**
- **j = 0,1,2,...n**

---

# 📊 Key Difference from Simple Linear Regression

| Simple Linear Regression | Multiple Linear Regression |
|-------------------------|----------------------------|
| One feature | Multiple features |
| hθ(x) = θ₀ + θ₁x | hθ(x) = θ₀ + θ₁x₁ + θ₂x₂ + ... |

---

# 🧠 Example

Predict house price using:

```
Price = θ₀ + θ₁(Rooms) + θ₂(Size) + θ₃(Location)
```

This allows the model to **capture more complex relationships**.

---

# 🎯 Summary

✔ Multiple Linear Regression uses **multiple input variables**

✔ Hypothesis function:

```
hθ(x) = θ₀ + θ₁x₁ + θ₂x₂ + ... + θₙxₙ
```

✔ Same **cost function (MSE)**

✔ Optimized using **Gradient Descent**
