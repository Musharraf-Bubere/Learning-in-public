# 🧠 1. Introduction to Linear Regression

Linear Regression is a **Supervised Machine Learning algorithm** used to predict **continuous numeric values**.

Examples:

- Predict house price
- Predict salary
- Predict height from weight

The goal is to find the **best-fit straight line** through the data points.

---

# 📊 Example Dataset

| Weight (X) | Height (Y) |
|------------|------------|
| 74 | 170 |
| 80 | 180 |
| 75 | 175 |

Where:

- **X → Independent variable (Input feature)**
- **Y → Dependent variable (Target / Output)**

Goal:

```
Predict height using weight
```

---

# 📈 Equation of a Straight Line

Basic mathematics equation:

```
y = mx + c
```

Where:

- **m → slope**
- **c → intercept**

Machine Learning form:

```
y = β₀ + β₁x
```

---

# 🤖 Hypothesis Function

In ML we call the prediction function **hypothesis**.

```
hθ(x) = θ₀ + θ₁x
```

Where:

| Parameter | Meaning |
|-----------|---------|
| θ₀ | Intercept |
| θ₁ | Slope (coefficient) |
| x | Input feature |

Prediction:

```
ŷ = hθ(x)
```

---

# 📉 Error

Difference between **actual value** and **predicted value**.

```
Error = y - ŷ
```

Where:

- **y → actual value**
- **ŷ → predicted value**

---

# 📊 Cost Function

To measure model performance we use **Mean Squared Error (MSE)**.

```
J(θ₀, θ₁) = 1/(2m) Σ (hθ(xᵢ) - yᵢ)²
```

Where:

- **m → number of training examples**
- **hθ(xᵢ) → predicted value**
- **yᵢ → actual value**

Goal:

```
Minimize the cost function
```

---

# 📉 Cost Function Graph

The cost function looks like a **bowl-shaped curve (convex)**.

Important point:

```
Global Minimum
```

At this point the **error is minimum**.

---

# ⚙️ Gradient Descent

Gradient Descent is an optimization algorithm used to find the **best values of θ₀ and θ₁**.

General update rule:

```
θⱼ = θⱼ − α (∂J / ∂θⱼ)
```

Where:

- **α → Learning rate**
- **∂J/∂θⱼ → Gradient (slope of cost function)**

Repeat this process until **convergence**.

---

# 🔄 Gradient Descent Update Equations

For **θ₀**:

```
θ₀ = θ₀ − α (1/m) Σ (hθ(xᵢ) − yᵢ)
```

For **θ₁**:

```
θ₁ = θ₁ − α (1/m) Σ (hθ(xᵢ) − yᵢ) xᵢ
```

---

# 📌 Learning Rate (α)

Learning rate controls **step size** of parameter updates.

| Learning Rate | Effect |
|---------------|-------|
| Too small | Slow learning |
| Too large | May overshoot minimum |
| Optimal | Fast convergence |

Example:

```
α = 0.001
```

---

# 🔄 Convergence

The algorithm repeats parameter updates until:

```
Cost function stops decreasing
```

This means the model reached **optimal parameters**.

---

```
Mean Squared Error (MSE)
```
