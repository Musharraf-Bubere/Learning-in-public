# 📊 Performance Metrics in Linear Regression

---

# 🧠 Why Performance Metrics Are Needed

After training a regression model, we need to measure:

```
How good the model predictions are
```

Performance metrics help evaluate the **accuracy of the regression model**.

Two important metrics used in Linear Regression:

1️⃣ R-Squared (R²)  
2️⃣ Adjusted R-Squared

---

# 1️⃣ R-Squared (Coefficient of Determination)

R² measures **how well the regression model explains the variance in the data**.

Formula:

```
R² = 1 - (SS_res / SS_total)
```

Where:

| Term | Meaning |
|-----|--------|
| SS_res | Residual Sum of Squares |
| SS_total | Total Sum of Squares |

---

# 📌 Expanded Formula

```
R² = 1 - [ Σ (yᵢ - ŷᵢ)² / Σ (yᵢ - ȳ)² ]
```

Where:

| Symbol | Meaning |
|------|------|
| yᵢ | Actual value |
| ŷᵢ | Predicted value |
| ȳ | Mean of actual values |

---

# 📉 Residual Error

Residual error is the difference between:

```
Actual value - Predicted value
```

```
Residual = yᵢ - ŷᵢ
```

The diagram in your mentor’s notes shows residual errors as the **vertical distance between the regression line and the data points**. :contentReference[oaicite:1]{index=1}

---

# 📊 Interpretation of R²

| R² Value | Meaning |
|------|------|
| 0 | Model explains none of the variance |
| 0.70 | Model explains 70% of variance |
| 0.85 | Model explains 85% of variance |
| 1 | Perfect model |

Example:

```
R² = 0.85
```

Means:

```
Model explains 85% of the data variance
```

---

# ⚠️ Problem with R-Squared

R² **always increases when new features are added**, even if the feature is useless.

Example dataset features:

```
Size of house
Number of bedrooms
Location
```

Adding more variables may increase **R² artificially**.

This can lead to **overfitting**.

---

# 2️⃣ Adjusted R-Squared

Adjusted R² solves the problem of R².

It penalizes the model for **adding unnecessary features**.

Formula:

```
Adjusted R² = 1 - [(1 - R²)(N - 1) / (N - P - 1)]
```

Where:

| Symbol | Meaning |
|------|------|
| N | Number of data points |
| P | Number of independent features |

---

# 📊 Example

If:

```
R² = 0.90
Features = 2
Adjusted R² ≈ 0.86
```

If more features are added:

```
R² = 0.92
Adjusted R² ≈ 0.82
```

This shows that **adding unnecessary features reduces Adjusted R²**.

---

# 🎯 Key Difference

| Metric | Behavior |
|------|------|
| R² | Always increases with more features |
| Adjusted R² | Increases only if feature improves model |

---

# 🧠 Final Summary

✔ R² measures **model goodness of fit**

✔ Formula:

```
R² = 1 - (SS_res / SS_total)
```

✔ Problem:

```
R² increases when features increase
```

✔ Solution:

```
Adjusted R²
```

✔ Adjusted R² penalizes unnecessary features
